```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')
```

# Finding a Local Minimum
Function 1: f(x,y)=x^2+y^2
*   local minimum: (0,0,0)


```
def f(x, y):
    return x ** 2 + y ** 2
  
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, 20, cmap='RdGy');
plt.plot([0.], [0.], markerfacecolor='k', markeredgecolor='k', marker='o', markersize=5, alpha=0.6)
```

```
def f(x, y):
    return x ** 2 + y ** 2

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot([0.], [0.], [0.], markerfacecolor='k', markeredgecolor='k', marker='o', markersize=5, alpha=0.6)

ax.view_init(60, 35)
```

# Finding an Inflection Point

Function 3: f(x,y) = -x^2 + y^2


*   inflection point: (0,0,0)



```
def f(x, y):
    return -x ** 2 + y ** 2
  
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, 20, cmap='RdGy');
plt.plot([0.], [0.], markerfacecolor='k', markeredgecolor='k', marker='o', markersize=5, alpha=0.6)
```
