# Original (y = w * x + b)

```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```
from mpl_toolkits import mplot3d
```

```
import numpy as np
np.random.seed(42)

xs = 2 * np.random.rand(100, 1)
ys = 4 + 3 * xs + np.random.rand(100, 1)

plt.plot(xs, ys, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

```
# split the data into training and test sets
# train set
train_xs = xs[:80]
train_ys = ys[:80]
# test set
test_xs = xs[80:]
test_ys = ys[80:]

# number of epochs
epochs = 10
# learning rate
lr = 0.01

# initial value for weight w and bias b
w = np.random.randn(1)
b = np.zeros(1)

for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w * train_xs[i] + b
    
    grad_w = (y_pred - train_ys[i]) * train_xs[i]
    grad_b = (y_pred - train_ys[i])
    
    w -= lr * grad_w
    b -= lr * grad_b
    
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w * test_xs[i] + b - test_ys[i]) ** 2
test_loss /= 20

test_loss
```

```
pred_ys = w * test_xs + b

plt.plot(test_xs, test_ys, "b.")
plt.plot(test_xs, pred_ys, "r.") # predicted values
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

# Using Two Weights (z = w1 * x + w2 * y + b)

```
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```
ax = plt.axes(projection='3d')

import numpy as np
np.random.seed(42)

# Data for three-dimensional scattered points
xdata = 2 * np.random.rand(100)
ydata = 2 * np.random.rand(100)
zdata = 4 + 3 * xdata + np.random.rand(100) + 3 * ydata + np.random.rand(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Dark2')
```

```
# split the data into training and test sets
# train set
train_Zs = zdata[:80]
train_Xs = xdata[:80]
train_Ys = ydata[:80]
# test set
test_Zs = zdata[80:]
test_Xs = xdata[80:]
test_Ys = ydata[80:]
```

```
# number of epochs
epochs = 10
# learning rate
lr = 0.01
```

```
# initial value for weight w and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
bias = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    z_pred = w1 * train_Xs[i] + w2 * train_Ys[i] + bias
    
    grad_w1 = (z_pred - train_Zs[i]) * train_Xs[i]
    grad_w2 = (z_pred - train_Zs[i]) * train_Ys[i]
    grad_bias = (z_pred - train_Zs[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    bias -= lr * grad_bias
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_Xs[i] + w2 * test_Ys[i] + bias - test_Zs[i]) ** 2
test_loss /= 20

test_loss
```

```
pred_Zs = w1 * test_Xs + w2 * test_Ys + b
```

```
ax = plt.axes(projection='3d')
ax.scatter3D(test_Xs, test_Ys, test_Zs, c=test_Zs, cmap='Blues')
ax.scatter3D(test_Xs, test_Ys, pred_Zs, c=pred_Zs, cmap='Reds')
```
